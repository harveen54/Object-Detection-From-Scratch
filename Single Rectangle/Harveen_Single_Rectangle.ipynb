```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import cv2
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
dimension = 8 # the dimensions of an image 8X8 image
minR = 0 # minimum index
maxR = dimension-1 # maximum index
max_deviation = 4 # maximum size square
num_imgs = 50000 # total number of images

def build_training_data():
    labels=[]
    dummy_images=[]
    for i in range(num_imgs):
        print('Processing {0} record\r'.format(i))
        img = np.ones((dimension,dimension)) * 255 #create a blank white image
        point_x_min = np.random.randint(0,maxR-2)

        x=min(point_x_min+4, maxR)
        point_x_max = np.random.randint(point_x_min+1 , x)
        point_y_min = np.random.randint(0, maxR-2)
        point_y_max = np.random.randint(point_y_min+1 , min(point_y_min+4, maxR))

        
        img[point_x_min:point_x_max+1,point_y_min:point_y_max+1] = 0 # marking square as black
        labels.append([point_x_min, point_y_min , point_x_max, point_y_max])
        #print(img.shape)

        cv2.imwrite('./dataset/'+str(i)+'.jpg', img)
        dummy_images.append(img)
    return dummy_images, labels
```

```python
raw_x, raw_y = build_training_data()
```

```python
def convert_and_save(raw_x, raw_y):
    X= np.array(raw_x)
    y= np.array(raw_y)
    np.save('./data', X)
    np.save('./labels',y)
    return X,y

def load_data():
    X = np.load('./data.npy')
    y = np.load('./labels.npy')
    return X,y
```

```python
X ,y = convert_and_save(raw_x, raw_y)
print(X.shape,y.shape)
```

```python

def manage_split(X,y, size=0.1, test_set=True):
    if(test_set):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, shuffle=True) #splitting into test and train datasets

        X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=size, shuffle=True) #splitting into valid and test datasets
        
        return X_train, y_train, X_valid, y_valid, X_test, y_test
    else:
        X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=size, shuffle=True) #splitting into test and train datasets
        
        return X_train, y_train, X_valid, y_valid


```

```python
X_train, y_train, X_valid, y_valid, X_test, y_test = manage_split(X, y)

print(X_train.shape,y_train.shape, X_valid.shape, y_valid.shape, X_test.shape, y_test.shape)
```

```python
# Define Model

from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten, Lambda
from keras.optimizers import SGD

model = Sequential([
        Lambda(lambda x: x / 255.0, input_shape =(8,8)),
        Flatten(),
        Dense(200), 
        Activation('relu'), 
        Dropout(0.2), 
        Dense(100), 
        Activation('relu'),
        Dropout(0.2),
        Dense(50), 
        Activation('relu'),
        Dense(25), 
        Activation('relu'),
        Dense(12), 
        Activation('relu'),
        Dense(4)
    ])

model.compile('adadelta', 'mse')
```

```python
model.fit(X_train, y_train, nb_epoch=10, validation_data=(X_valid, y_valid), verbose=1)
```

```python
# Inference Time

def IOU(bbox1, bbox2):
    '''Calculate overlap between two bounding boxes [x, y, w, h] as the area of intersection over the area of unity'''
#     x1, y1, w1, h1 = bbox1[0], bbox1[1], bbox1[2], bbox1[3]
#     x2, y2, w2, h2 = bbox2[0], bbox2[1], bbox2[2], bbox2[3]

    x_min_1, y_min_1, x_max_1, y_max_1 = bbox1[0], bbox1[1], bbox1[2], bbox1[3]
    x_min_2, y_min_2, x_max_2, y_max_2 = bbox2[0], bbox2[1], bbox2[2], bbox2[3]

    w_I = min(x_max_1, x_max_2) - max(x_min_1, x_min_2)
    h_I = min(y_max_1, y_max_2) - max(y_min_1, y_min_2)
   
    if w_I <= 0 or h_I <= 0:  # no overlap
        return 0.
    
    I = w_I * h_I

    U = (x_max_1-x_min_1) * (y_max_1-y_min_1) + (x_max_2-x_min_2) * (y_max_2-y_min_2) - I

    return I / U
```

```python
iou = []
for index, item in enumerate(zip(X_test, y_test)):
    out = model.predict(item[0].reshape(1,8,8))
    x_min, y_min, x_max, y_max = out[0]
    iou.append(IOU([x_min, y_min, x_max, y_max] , item[1]))

numpy_iou = np.array(iou)
    #print(iou)
```
